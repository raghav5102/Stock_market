```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('RELIANCE.NS.csv',parse_dates = ['Date'] )
df.head()
```

```python
series = df['Close']
```

```python
series
```

```python
series1 = df['Close'].rolling(window = 15).mean()
```

```python
for i in range(0,14):
    series1[i] = series[i]
series1
```

```python
df['moving_average_15'] = series1
```

```python
df.head(20)
```

```python
series2 = df['Close'].rolling(30).mean()
```

```python
for i in range(0,29):
    series2[i] = series[i]
series2
```

```python
df['moving_average_30'] = series2
```

```python
df.head(35)
```

```python
series3 = df['Close'].rolling(60).mean()
```

```python
for i in range(0,59):
    series3[i] = series[i]
series3
```

```python
df['moving_average_60'] = series3
```

```python
df.head(60)
```

```python
#Calculating RSi
df['difference'] = df['Close'].diff()
```

```python
df.head()
```

```python
df['up'] = df['difference'].clip(lower = 0)
```

```python
df.head()
```

```python
df.head(30)
```

```python
df['down'] = -1*df['difference'].clip(upper = 0)
```

```python
df.head()
```

```python
ema_up = df['up'].ewm(com=13, adjust=False).mean()
```

```python
ema_down = df['down'].ewm(com=13, adjust=False).mean()
```

```python
rs = ema_up/ema_down
```

```python
df['RSI'] = 100 - (100/(1 + rs))
```

```python
df.head()
```

```python
df.head(60)
```

```python
df.drop(['up','down','difference'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
value = df['RSI'].mean()
value
```

```python
df['RSI'] = df['RSI'].fillna(value = value)
```

```python
df.head()
```

```python
#Calculating Bollinger Band

df['15_MA'] = df['Close'].rolling(window = 15).mean()
df['15_SD'] = df['Close'].rolling(window = 15).std()
```

```python
df.head()
```

```python
df['Bollinger_Upper'] = df['15_SD'] + 2*df['15_SD']
df['Bollinger_Lower'] = df['15_SD'] - 2*df['15_SD']
```

```python
df.head(15)
```

```python
value1 = df['Bollinger_Upper'].mean()
```

```python
df['Bollinger_Upper'] = df['Bollinger_Upper'].fillna(value= value1)
```

```python
df.head()
```

```python
value2 = df['Bollinger_Lower'].mean()
```

```python
df['Bollinger_Lower'] = df['Bollinger_Lower'].fillna(value= value2)
```

```python
df = df.drop(['15_MA','15_SD'],axis =1)
```

```python
df.head()
```

```python
df.to_csv('Final_Dataset.csv')
```
